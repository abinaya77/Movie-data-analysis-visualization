```python
#fill missing values with zeros values 


df.fillna(0, inplace = True)

#checking if nan values are filled
count = df.isna().sum()
print ("NAN count in series:\n ", count)
```

```python
#already revenue in millions so no prob if in $ divide the column by 1000000 to convert to millions
####   df["revenue"]=df["revenue"]/1000000
```

```python
#top10 movies which created highest revenue


# Sort the dataframe with the 'revenue' column as reference using the 'sort_values' function. Make sure to set the argument
#'ascending' to 'False'

df=df.sort_values(by="Revenue (Millions)",ascending=False)

# Get the top 10 profitable movies by using position based indexing. Specify the rows till 10 (0-9)

df.iloc[:10,:]

```

```python
# Change the scale of Metascore from 100 to 10 (as ratings is also in 10 scale )

df["Metascore"]=df["Metascore"]/10

df.head()

```

```python
# Find the average ratings which is avg btw ratings and metascore and store in sep column 

df["Avg_rating"]=(df["Rating"]+df["Metascore"])/2
df

```

```python
#Sort in descending order of average rating
df=df[["Title","Rating","Metascore","Avg_rating"]]
df=df.loc[abs(df["Rating"]-df["Metascore"]<0.5)]

# Find the movies with metacritic-IMDb_rating < 0.5 and also with the average rating of >8
df=df.sort_values(by="Avg_rating",ascending=False)
New_df=df.loc[df["Avg_rating"]>=8]
New_df

	#note :
 ## after running this commands reset index or errors will arise ..or read the data from first 
```

```python
###task -1
###which genre has got highest scores based on votes,revenue,metascore,avg_rating..
```

```python
# Create the dataframe df_by_genre
df_by_genre=df.loc[:,"Rating":"Avg_rating"]
df_by_genre[["Genre","Genre_1","Genre_2"]]=df[["Genre","Genre_1","Genre_2"]]

df_by_genre

```

```python
# Create a column cnt and initialize it to 1
df_by_genre["cnt"]=1

df_by_genre[["Genre","Genre_1","Genre_2"]]

```

```python
# Group the movies by individual genres
df_by_g1=df_by_genre.groupby("Genre").aggregate(np.sum)
df_by_g2=df_by_genre.groupby("Genre_1").aggregate(np.sum)
df_by_g3=df_by_genre.groupby("Genre_1").aggregate(np.sum)

df_by_g1

```

```python
df_by_g2
```

```python
# Add the grouped data frames and store it in a new data frame
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```

```python
# Extract genres with atleast 10 occurences
genre_top_10=df_add.loc[df_add["cnt"]>10]

genre_top_10

```

```python
# Take the mean for every column by dividing with cnt 
genre_top_10.iloc[:,0:-1]=genre_top_10.iloc[:,0:-1].divide(genre_top_10["cnt"],axis=0)

genre_top_10

```

```python
# Rounding off the columns of Votes to two decimals
genre_top_10.loc[:,"Rating":"Avg_rating"]=round(genre_top_10.loc[:,"Rating":"Avg_rating"],2)
genre_top_10
```

```python
# Countplot for genres
sns.barplot(x=genre_top_10["cnt"],y=genre_top_10.index)
plt.show()
```
